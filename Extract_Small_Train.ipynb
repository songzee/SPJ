```python
import pandas as pd
import json
```

```python
with open('train_ids.json', 'r') as f:
    train_id = json.load(f)
```

```python
train_id_small = train_id[0:100]
```

```python
with open('train.json', 'r') as f:
    train = json.load(f)
```

```python
train['v_QOlSCBRmfWY']
```

```python
columns=['id','duration','sentences','timestamps']
df = pd.DataFrame(columns=columns)
```

```python
for id in train_id_small:
    if id in train.keys():
        new = pd.DataFrame({'id': id,'duration':train[id]['duration'],'sentences':train[id]['sentences'],'timestamps':train[id]['timestamps']})
        df = pd.concat([df,new])
```
